```python
import speech_recognition as sr
from gtts import gTTS
import pyttsx3 as p
from ipynb.fs.full.web_auto_wiki import *
from ipynb.fs.full.web_auto_movie import *
from ipynb.fs.full.web_auto_music import *
from ipynb.fs.full.web_auto_recomm import *
from ipynb.fs.full.words import *

r = sr.Recognizer()
engine = p.init()
engine.say("Hello user, how are you doing?")
engine.runAndWait()
with sr.Microphone() as source:
    audio_data = r.listen(source)
    print("Recognizing...")
    try:
        text = r.recognize_google(audio_data)
        print(text)
    except:
        print("Sorry could not recognize your voice, please try again")

#cue for giving instructions
engine.say("what you would like to do?")
engine.runAndWait()
print("what do you want")

#giving instructions
r2 = sr.Recognizer()
with sr.Microphone() as source6:
    audio_data2 = r2.listen(source6)
    print("Recognizing...")
    try:
        instruction = r2.recognize_google(audio_data2)
        print(instruction)
    except:
        print("Sorry could not recognize your voice, please try again")

#getting info from wiki:
r3 = sr.Recognizer()
if "information" in instruction:
    engine.say("information about what?")
    engine.runAndWait()
    with sr.Microphone() as source1:
        audio_data3 = r3.listen(source1)
        print("Recognizing...")
        try:
            information = r3.recognize_google(audio_data3)
            bot = info()
            bot.get_info(information)
        except:
            print("Sorry could not recognize your voice, please try again")
    
#movie rating
r4 = sr.Recognizer()
if "review" in instruction:
    engine.say("what is the name of the movie?")
    engine.runAndWait()
    with sr.Microphone() as source2:
        r4.adjust_for_ambient_noise(source2)
        audio_data4 = r4.listen(source2)
        print("Recognizing...")
        try:
            rating = r4.recognize_google(audio_data4)
            print(rating)
            bot = Movie()
            bot.movie_review(rating)
        except:
            print("Sorry could not recognize your voice, please try again")

#playing music
r5 = sr.Recognizer()
if "music" in instruction:
    engine.say("which artist should I play?")
    engine.runAndWait()
    with sr.Microphone() as source3:
        audio_data5 = r5.listen(source3)
        print("Recognizing...")
        try:
            video = r5.recognize_google(audio_data5)
            print(video)
            bot = music()
            bot.play(video)
        except:
            print("Sorry could not recognize your voice, please try again")

#movie recommendation       
if "recommendation" in instruction:
    engine.say("Here is the list of movies for you")
    engine.runAndWait()
    bot = recom()
    bot.recom_info()
    
#getting meaning of a word     
r6 = sr.Recognizer()
if "meaning" in instruction:
    engine.say("which word sire?")
    engine.runAndWait()
    with sr.Microphone() as source4:
        audio_data6 = r6.listen(source4)
        print("Recognizing...")
        try:
            word = r6.recognize_google(audio_data6)
            print(word)
            mean(word)
        except:
            print("Sorry could not recognize your voice, please try again")
        
        
#translate a word     
r7 = sr.Recognizer()
if "translate" in instruction:
    engine.say("which word sire?")
    engine.runAndWait()
    with sr.Microphone() as source5:
        audio_data7 = r7.listen(source5)
        print("Recognizing...")
        try:
            word1 = r7.recognize_google(audio_data7)
            print(word1)
            translate(word1)
        except:
            print("Sorry could not recognize your voice, please try again")
```
